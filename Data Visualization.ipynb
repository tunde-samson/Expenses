### Using plotly for Age VS Total Income

```python
ex.bar(user_data, x='age', y='total_income', color='age')
```

### Using seaborn for Age VS Total Income

```python
plt.figure(figsize=(15,8))
sns.barplot(user_data, x='age', y='total_income', color='salmon')
plt.title('Total Income in respect to Age')
plt.xticks(rotation=90);

plt.savefig('Age_vs_Total_Income.png')

#to save the image or chart i can just right click save image on desktop then copy to powerpoint
#or using plt.savefig() which will save to the working directory
```
